# pie 함수로 원그래프 그리기

```python
import matplotlib.pyplot as plt

data = [10, 20, 30, 40]
name = ['A형', 'B형', 'AB형', 'O형']

plt.rc('font', family='Malgun Gothic')
plt.pie(data, labels=name)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

data = [15, 25, 35, 45]
name = ['A형', 'B형', 'AB형', 'O형']

plt.rc('font', family='Malgun Gothic')
plt.pie(data, labels=name, autopct='%.3f%%')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

data = [15, 25, 35, 45]
name = ['A형', 'B형', 'AB형', 'O형']

plt.rc('font', family='Malgun Gothic')
plt.pie(data, labels=name, autopct='%.3f%%')
plt.legend(loc=3)
plt.show()
```

```python
import matplotlib.pyplot as plt

data = [15, 25, 35, 45]
name = ['A형', 'B형', 'AB형', 'O형']

plt.rc('font', family='Malgun Gothic')
plt.pie(data, labels=name, autopct='%.2f%%', colors=['violet', 'skyblue', 'yellow', 'm'], explode=(0, 0, 0.1, 0))
plt.legend(loc=3)
plt.show()
```

```python
import matplotlib.pyplot as plt

data = [15, 25, 35, 45]
name = ['A형', 'B형', 'AB형', 'O형']

plt.rc('font', family='Malgun Gothic')
plt.pie(data, labels=name, autopct='%.2f%%', colors=['violet', 'skyblue', 'yellow', 'm'], explode=(0, 0, 0.07, 0), startangle=90)
plt.legend(loc=1)
plt.show()
```

# scatter()

```python
import matplotlib.pyplot as plt

plt.scatter([1, 2, 3, 4], [10, 30, 20, 41])
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.scatter([1, 2, 3, 4], [10, 30, 20, 41], s=[90, 180, 225, 360])
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.scatter([1, 2, 3, 4], [10, 30, 20, 41], s=[90, 180, 225, 360], c=['red', 'skyblue', 'k', 'gold'])
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.scatter([1, 2, 3, 4], [10, 30, 20, 41], s=[90, 180, 225, 360], c=range(4), cmap='rainbow')
plt.grid(True)
plt.colorbar()
plt.title('scatter')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import matplotlib.pyplot as plt
import random

x = []
y = []
size = []

for i in range(100):
    x.append(random.randint(50, 100))
    y.append(random.randint(50, 100))
    size.append(random.randint(10, 100))

plt.scatter(x, y, s=size)
plt.grid(True)
plt.title('random point')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import matplotlib.pyplot as plt
import random

x = []
y = []
size = []

for i in range(100):
    x.append(random.randint(50, 100))
    y.append(random.randint(50, 100))
    size.append(random.randint(10, 100))

plt.scatter(x, y, s=size, c=size, cmap='rainbow')
plt.grid(True)
plt.title('random point')
plt.xlabel('x')
plt.ylabel('y')
plt.colorbar()
plt.show()
```

```python
import matplotlib.pyplot as plt
import random

x = []
y = []
size = []
sizes = []

for i in range(100):
    x.append(random.randint(50, 100))
    y.append(random.randint(50, 100))
    size.append(random.randint(10, 100))
    sizes.append(random.randint(10, 300))

plt.scatter(x, y, s=size, c=sizes, cmap='rainbow', alpha=0.81)
plt.grid(True)
plt.title('random point')
plt.xlabel('x')
plt.ylabel('y')
plt.colorbar()
plt.show()
```
