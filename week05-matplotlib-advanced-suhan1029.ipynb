## figsize

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(7, 3)) # 가로길이, 세로길이
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.plot([-2, -1, 0, 1, 2, 3], [20, 30, 15, 40, 45, 35], label='선 1')
plt.plot([-2, -1, 0, 1, 2, 3], [-25, 10, 35, 20, 15, 25], label='선 2')
plt.title('선 그래프 예제')
plt.xlabel('x축')
plt.ylabel('y축')
plt.legend()
plt.show()
```

## boxplot

```python
import random as rd
import matplotlib.pyplot as plt

result = []

for i in range(13):
    result.append(rd.randint(1, 1000))

print(sorted(result))

plt.boxplot(result)
plt.show()
```

```python
import random as rd
import matplotlib.pyplot as plt

result1 = [rd.randint(1, 500) for i in range(13)]
result2 = [rd.randint(501, 1000) for i in range(13)]

plt.boxplot(result1)
plt.boxplot(result2)
plt.show()
```

```python
import random as rd
import matplotlib.pyplot as plt

result1 = [rd.randint(1, 500) for i in range(13)]
result2 = [rd.randint(501, 1000) for i in range(13)]

plt.boxplot([result1, result2], whis=0.7)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# np.random.seed(10)

f1 = plt.figure(figsize=(10, 2))
plt.plot(np.random.randn(100))

plt.title("figure size : (10, 2)")
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.subplot(2, 3, 1)
plt.subplot(2, 3, 5)
plt.subplot(2, 3, 6)
```

```python
import numpy as np
import matplotlib.pyplot as plt

ax1 = plt.subplot(2, 1, 1)
plt.plot(np.random.randint(1, 10, 10), 'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')
print(ax1)

ax2 = plt.subplot(2, 1, 2)
plt.plot(np.arange(10), np.arange(10)*2, 'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')
print(ax2)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2)

axes[0, 0].plot([1, 2, 3, 4])
axes[0, 0].set_title("Subplot 1")

axes[0, 1].plot([4, 3, 2, 1])
axes[0, 1].set_title("Subplot 2")

axes[1, 0].hist([1, 2, 2, 3, 3, 3, 4, 4, 4, 4], bins=5, rwidth=0.75)
axes[1, 0].set_title("Subplot 3")

axes[1, 1].scatter([1, 2, 3, 4], [4, 3, 2, 1])
axes[1, 1].set_title("Subplot 4")

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2)

# np.random.seed(0)
axes[0, 0].plot(np.random.rand(5))
axes[0, 0].set_title("axes 1")
axes[0, 1].plot(np.random.rand(5))
axes[0, 1].set_title("axes 2")
axes[1, 0].plot(np.random.rand(5))
axes[1, 0].set_title("axes 3")
axes[1, 1].plot(np.random.rand(5))
axes[1, 1].set_title("axes 4")

plt.tight_layout()
plt.show()
```
