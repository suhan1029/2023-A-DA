# 판다스로 구현하는 연령별 인구구조 유사지역

```python
import pandas as pd

df = pd.read_csv('age.csv', encoding='cp949', index_col=0)
df.head()
```

```python
df = df.replace(',', '', regex=True)
df.head()
```

```python
df = df.apply(pd.to_numeric)
df.dtypes
```

```python
df = df[df[df.columns[0]] != 0]
```

```python
df = df.div(df[df.columns[0]], axis='index')
df.head(4)
```

```python
del df['2023년09월_계_총인구수'], df['2023년09월_계_연령구간인구수']
df.head(4)
```

```python
name = input('원하는 지역의 이름을 입력해주세요: ')
a = df.index.str.contains(name)
df2 = df[a]
df2
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
df2.T.plot(rot=45)
plt.grid()
plt.show()
```

```python
x = df.sub(df2.iloc[0], axis='columns') # axis=1
x.head(4)
```

```python
import numpy as np

y = np.power(x, 2)
z = y.sum(axis=1)
z
```

```python
i = z.sort_values().index[:6]
i
```

```python
df.loc[i].T.plot(figsize=(12, 7))
plt.xlabel('\n나이')
plt.ylabel('인구 비율\n')
plt.grid()
plt.show()
```

```python
x = np.arange(1, 5, .5)
y = np.random.randint(1, 5, 8)
size = [50, 80, 100, 20, 150, 200, 300, 250]
plt.scatter(x, y, s=size, c=size, cmap='jet')
plt.xlim(0, 6)
plt.ylim(0, 6)
plt.colorbar()
plt.grid()
plt.show()
```
