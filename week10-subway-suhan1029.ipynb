```python
import csv

f = open('subwayfee.csv', encoding='UTF-8')
data = csv.reader(f)

header = next(data)
print(header)

rate = 0
max_rate = 0

for row in data:
    for i in range(4, 8):
        row[i] = int(row[i].replace(',', ""))
    
    if (row[6] != 0) & (row[4] + row[6] > 100000):
        rate = row[4] / (row[4] + row[6])
        if rate > max_rate:
            max_rate = rate
            print(row, round(max_rate, 2))

f.close()
```

```python
import csv
import numpy as np

f = open('subwayfee.csv', encoding='UTF-8')
data = csv.reader(f)

header = next(data)

max_people = [0] * 4
max_station = ['', '', '', '']
label = ['(유임승차)', '(유임하차)', '(무임승차)', '(무임하차)',]

for row in data:
    for i in range(4, 8):
        row[i] = int(row[i].replace(',', ""))
        if row[i] > max_people[i-4]:
            max_people[i-4] = row[i]
            max_station[i-4] = row[3] + ' ' + row[1]

f.close()

for i in range(4):
    print(f'{label[i]} {max_station[i]}: {max_people[i]}명')
```

```python
import csv
import matplotlib.pyplot as plt

f = open('subwayfee.csv', encoding='UTF-8')
data = csv.reader(f)

header = next(data)
label = ['(유임승차)', '(유임하차)', '(무임승차)', '(무임하차)',]
color = ['skyblue', 'palegreen', 'pink', 'beige']

station = input('지하철역 이름 입력: ')

plt.rc('font', family='Malgun Gothic')
for row in data:
    if row[3] == station: # 특정한 하나의 지하철역만 출력()
        for i in range(4, 8):
            row[i]  = int(row[i].replace(',', ""))
        plt.title(row[3] + ' ' + row[1])
        plt.pie(row[4:8], labels=label, colors=color, autopct='%1.f%%')
        plt.legend()
        plt.savefig(row[3] + ' ' + row[1] + '.png') # 보고 싶은 지하철역의 정보를 저장
        plt.show()

f.close()
```
