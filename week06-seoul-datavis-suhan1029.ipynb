```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)

header = next(data)

mean = []
low = []
high = []

for row in data:
    if (row[2] != '') & (row[3] != '') & (row[4] != ''):
        if (row[0].split('-')[1] == '12') & (row[0].split('-')[2] == '25'):
            mean.append(float(row[2]))
            low.append(float(row[3]))
            high.append(float(row[4]))



f.close()

plt.figure(figsize=(13, 2))
plt.plot(mean, 'k')
plt.plot(low, 'b')
plt.plot(high, 'r')
plt.xlabel('Day (December 25)')
plt.ylabel('Temperture(*C)')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)

header = next(data)

mean = []
low = []
high = []

for row in data:
    if (row[2] != '') & (row[3] != '') & (row[4] != ''):
        if (row[0].split('-')[1] == '12') & (row[0].split('-')[2] == '25'):
            mean.append(float(row[2]))
            low.append(float(row[3]))
            high.append(float(row[4]))



f.close()

plt.figure(figsize=(13, 4))
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title('크리스마스의 기온 변화 그래프')
plt.plot(mean, 'k', label='mean')
plt.plot(low, 'b', label='low')
plt.plot(high, 'r', label='high')
plt.xlabel('Day (December 25)')
plt.ylabel('Temperture(*C)')
plt.legend()
plt.grid()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)

header = next(data)

temp_day = []
temp_mean = []
year = []

for row in data:
    if int(row[0].split('-')[0]) > 1907:
        if row[2] != '':
            temp_day.append(float(row[2]))
            if (row[0].split('-')[1] == '12') & (row[0].split('-')[2] == '31'):
                year.append(int(row[0].split('-')[0]))
                temp_mean.append(sum(temp_day) / len(temp_day))
                temp_day = []



f.close()

plt.figure(figsize=(13, 4))
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title('크리스마스의 기온 변화 그래프')
plt.plot(year, temp_mean, color='b', marker='d')
plt.xlabel('Year')
plt.ylabel('Average Temperture')
plt.grid()
plt.show()
```

#  기온 데이터를 다양하게 시각화하기

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
header = next(data)
result = []

for row in data:
    if row[4] != '':
        result.append(float(row[4]))

plt.hist(result, bins=100, color='r')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
header = next(data)
aug = []

for row in data:
    month = row[0].split('-')[1]
    if row[4] != '':
        if month == '08':
            aug.append(float(row[4]))

plt.hist(aug, bins=100, color='r')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
header = next(data)
jan = []
aug = []

for row in data:
    month = row[0].split('-')[1]
    if row[4] != '':
        if month == '01':
            jan.append(float(row[4]))
        elif month == '08':
            aug.append(float(row[4]))

plt.hist(jan, bins=100, color='b', label='January')
plt.hist(aug, bins=100, color='r', label='August')
plt.legend()
#plt.grid()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
header = next(data)

month = [[], [], [], [], [], [], [], [], [], [], [], []]

for row in data:
    if row[4] != '':
        month[int(row[0].split('-')[1])-1].append(float(row[4])) # 0번째 인덱스에 1월의 데이터를 넣어야 하므로 -1한다.

plt.boxplot(month)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
header = next(data)

day = []
for i in range(31):
    day.append([])

for row in data:
    if row[4] != '':
        if row[0].split('-')[1] == '01':
            day[int(row[0].split('-')[2])-1].append(float(row[4]))

plt.style.use('ggplot')
plt.figure(figsize=(14, 6))
plt.boxplot(day, showfliers=False)
plt.show()
```
